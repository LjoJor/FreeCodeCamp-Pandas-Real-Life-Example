```python
btc_o = pd.read_csv("btc-market-price.csv",
            index_col=0,
            names = ["Date", "BTC"],
            parse_dates=True)
btc_o.head()
```

```python
btc = btc_o.copy()
len(btc)
```

###Plotting basics

```python
btc.plot(figsize = (10,6))
```

```python
eth_df = pd.read_csv("eth-price.csv",
            header = 0,
            names = ["Date","Timestamp","ETH"],
            index_col=0,
            parse_dates = True,
            usecols=[0,2] 
            )
eth_df.head()
len(eth_df)
```

```python
prices = pd.DataFrame(index= btc.index)
prices.head(
)
prices["Bitcoin"] = btc["BTC"]
prices["Ether"] = eth_df["ETH"]
prices.head()
```

```python
prices.plot(figsize =(16,10))
```
